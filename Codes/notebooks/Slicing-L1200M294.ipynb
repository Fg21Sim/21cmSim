```python
def batch_write_slice( FgSimLightCone, fzlist, clobber_=False ):
    r"""
        *** Batch write slices *** # Basic function
        !!!  !!! # Important note
        +++  +++ # Improvements
        :Params FgSimLightCone: a FgSimLightCone lightcone object;
        :Params fzlist: a fzlist dict from genfz();
        :Params clobber_: clobber handle
    """
    outfile_pattern = "{prefix}_f{f:06.2f}_z{z:06.3f}.fits"
    outfile_prefix = "deltaTb"
    
    for z, f in zip( fzlist['z'], fzlist['f']):
        outfile = outfile_pattern.format(prefix = outfile_prefix, z=z, f=f)
        logger.info("z=%06.3f, freq=%06.2f MHz : %s ..." % (z, f, outfile))
        data = FgSimLightCone.get_slice(z)
        FgSimLightCone.write_slice(outfile, data=data, z=z, freq=f, clobber=clobber_)
```

```python
batch_write_slice(Slc, fzlist1)
```
